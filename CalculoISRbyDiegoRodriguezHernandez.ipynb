##Definimos el limite y el porcentaje ISR

````python
limiteInf = 18837.38
porc = 0.2352
````

````python
##Función de calcular el ISR mensual
def calISRmensual(sueldo):
    exc = sueldo - limiteInf #calculamos el excedente
    ##estructura utilizada
    if exc > 0: #si el excedente es mayor a 0, hacemos el isr
       isr = exc * porc
    else: #de lo contrario el isr es 0
      isr = 0
    return round (isr,2) #redondeamos a 2 decimales

##calculamos el ISR MENSUAK
df["ISRMes"] = df["Enero"].apply(calISRmensual)

##calculamos el anual
df["ISRAnual"] = df["ISRMes"] * 12
````
